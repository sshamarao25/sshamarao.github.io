### b. The Simple Classification Network

Now construct the network. Fill in your code for the __init__ and forward methods. Again, we want to have two hidden layers (dims: hidden_dim_1, hidden_dim_2) with relu activation functions, and output layer of dimension n_classes (and softmax activation function). The model should return both the probabilities (probs) and the logits (logits), as you can tell from the return statement.

```python
class SimpleClassificationNertwork(nn.Module):
    def __init__(self, input_dim, hidden_dim_1, hidden_dim_2, n_classes):
        super().__init__()
        ### YOUR CODE HERE
        self.layers = nn.Sequential (
            # First hidden layer
            nn.Linear(input_dim, hidden_dim_1),
            # Activating Relu
            nn.ReLU(),
            # Second hidden layer
            nn.Linear(hidden_dim_1, hidden_dim_2),
            # Activating Relu
            nn.ReLU(),
            # Output layer
            nn.Linear(hidden_dim_2, n_classes),
            # Activating Softmax on the final layer
            nn.Softmax(dim=1)
        )

        ### END YOUR CODE

    def forward(self, x):
        ### YOUR CODE HERE
        probs = self.layers(x)
        logits = torch.log(probs)


        ### END YOUR CODE
        return probs, logits
```

```python
mySimpleClassificationNertwork = SimpleClassificationNertwork(input_dim=input_dim,
                                                              hidden_dim_1=7,
                                                              hidden_dim_2=10,
                                                              n_classes=n_classes)
```

Now verify that this loss agrees with the manual calculation. Recall from Week 2 that

$$ CE \rightarrow -\frac{1}{N}\sum_k \log(q^k_{{correct \ class}}) $$

where k refers to the example, N is the number of examples, and

$$q_{{correct \ class}}$$

is the model probability for the correct class for a given example.

```python
# call your manual loss calculation manual_loss

### YOUR CODE HERE
# Correct classes are - 0,1,2,0 for the 4 examples
# To calculate manual loss, first we need to calculate probability for each class
sum_log_probs = 0
classes_ex = [0,1,2,0]

for i in range(logits.shape[0]):
  # First let us get the logits for each class
  logits_1 = logits[i]
  # Convert the logits into exponential
  logits_exp = np.exp(logits_1.detach().numpy())
  # Calculate the sum
  logits_sum = np.sum(logits_exp)
  # Calculate the probability
  p_1_1 = torch.tensor(logits_exp/logits_sum)
  # Calculate the log of probability
  log_1 = torch.log(p_1_1[classes_ex[i]])
  # Sum it up as we loop through classes
  sum_log_probs += log_1

# Calculate the Average loss
manual_loss = -sum_log_probs/len(classes_ex)

### END YOUR CODE
manual_loss
```

Now apply the gpt2_model to a sentence in order to predict the most likely next word after 'The movement started in Italy. From there it went to France and Switzerland.  Soon it spread throughout'. You may want to consult the PyTorch Introduction II Notebook and the Week 2 lesson notebook.

Please get i) the shape of the output, ii) the values of the logits of the last token,  iii) the index of the  token with largest logit, and iv) the token that belongs to it. Same for the token with the second most largest logit.

```python
text = "The movement started in Italy. From there it went to France and Switzerland.  Soon it spread throughout"
input = gpt_2_tokenizer(text, return_tensors='pt')

gpt_out = gpt2_model(**input)

# Please call your model output shape, output_shape, the logits for the last position output_logits,
# the index for the token with the largest output_logit value max_logit_index, and the corresponding token max_logit_token.
# Name the corresponding values for the second largest logit second_logit_index, second_logit_token, and second_logit.

### YOUR CODE HERE
output_shape = gpt_out.logits.shape
output_logits = gpt_out.logits

# Lesson 2 guidance
last_token_logits = (output_logits.detach()[0, -1])
last_token_indices = np.argsort(last_token_logits)
top_logits, top_logit_indices = torch.topk(last_token_logits, k=2)

max_logit_index = top_logit_indices[0].item()
max_logit = top_logits[0].item()
max_logit_token = gpt_2_tokenizer.decode([max_logit_index])

second_logit_index = top_logit_indices[1].item()
second_logit = top_logits[1].item()
second_logit_token = gpt_2_tokenizer.decode([second_logit_index])

### END YOUR CODE

print('Output shape: ', output_shape)
print('Logits of output for last token: ', output_logits)
print('Index of token with largest logit: ', max_logit_index)
print('Token with largest logit: ', max_logit_token)
print('Logit of token with largest logit: ', max_logit)
print('Index of token with second largest logit: ', second_logit_index)
print('Token with second largest logit: ', second_logit_token)
print('Logit of token with second largest logit: ', second_logit)
```

Now we will translate the logits into relative token probabilities depending on the chosen temperature. Use numpy or pytorch calculations. (But remember to use .detach() etc if you want to use numpy.)

```python
T_1 = 1.
T_2 = 10.
T_3 = 0.1

# Please call your relative probabilities between the most likely token and the second most likely token p_t1, p_t2, p_t3, depending
# on the temperature

### YOUR CODE HERE
likely_logits = torch.tensor([[max_logit, second_logit]])

# Create a temp list to hold all the temperature and probability of scaled logits
temp_list = [T_1, T_2, T_3]
p_scaled_logits = []

# Calculation includes e^logit(i)/T/Sum of j(e^logit(j)/T)
# 1. For each of the logit, calculate the e^logit(i)/T
# convert likely_logits into a numpy array
for t in range(len(temp_list)):
  likely_scaled_logits = (likely_logits/temp_list[t]).detach().numpy()

  # Find the max val in the same axis
  max_val = np.max(likely_scaled_logits, axis=-1, keepdims=True)

  # To prevent overflow, subtrack from max val before calculating exponential
  scaled_logits = np.exp(likely_scaled_logits - max_val)

  # 2. Calculate the probability by dividing each scaled logit from the sum
  sum_scaled_logits = np.sum(scaled_logits)
  p_scaled_logits.append((scaled_logits/sum_scaled_logits))

# Convert the list into tensor
p_scaled_logits = torch.tensor(p_scaled_logits)
p_t1 = p_scaled_logits[0]
p_t2 = p_scaled_logits[1]
p_t3 = p_scaled_logits[2]

# Using pytorch to calculate probability - used for comparison with the manual calc
# p_t1 = torch.softmax(likely_logits/T_1, dim=1)
# p_t2 = torch.softmax(likely_logits/T_2, dim=1)
# p_t3 = torch.softmax(likely_logits/T_3, dim=1)

# ### END YOUR CODE

print('Logit ratio for T1: ', p_t1)
print('Logit ratio for T2: ', p_t2)
print('Logit ratio for T3: ', p_t3)
```
